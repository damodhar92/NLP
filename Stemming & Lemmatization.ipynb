### Stemming 

* In natural language processing (NLP), stemming is a text normalization process that involves reducing words to their base or root form. The goal of stemming is to simplify words to their common base or root, so that variations of a word are treated as the same word. 

* spacy library is not supported for stemming so we can use NLTK

```python
import nltk
import spacy
```

```python
from nltk.stem import PorterStemmer

stemmer=PorterStemmer()
```

```python
words=['eating','eats','eat','ate','adjustable','rafting','ability','meeting']

for word in words:
    print(word, "|", stemmer.stem(word))
    
```

### Lemmatization

* Lemmatization is a linguistic and natural language processing (NLP) technique that involves reducing words to their base or root form, known as the lemma. The lemma is the canonical, dictionary form of a word.

```python
#lemetizatin in spacy
nlp=spacy.load('en_core_web_sm')
doc=nlp("eating eats eat ate adjustable rafting ability meeting better")

for token in doc:
    print(token, "|", token.lemma_)
```

```python
doc=nlp("Hai my name is damodhara iam currently learning nlp")

for token in doc:
    print(token, "|", token.lemma_)
```

```python
doc=nlp("mando taked for 3 hours although taking isn't this he became talkative")

for token in doc:
    print(token, "|", token.lemma_)
```

```python
ar = nlp.get_pipe("attribute_ruler")
ar.add([[{"TEXT":"Bro"}],[{"TEXT":"Brah"}]],{"LEMMA":"Brother"})
doc=nlp("Bro, you wanna go? Brah, don't say no! I am exhausted")

for token in doc:
    print(token.text, "|", token.lemma_)
```
